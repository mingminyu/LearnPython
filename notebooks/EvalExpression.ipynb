# 使用 eval 一行代码实现阶乘

```python
def factorial(n):
    res = 1
    for i in range(1, n + 1):
        res *= i

    return res

factorial(3)
```

```python
def factorial_eval(n):
    return eval("*".join([str(i + 1) for i in range(n)]))

factorial_eval(3)
```

`math` 库中的 `factorial` 本身就实现了阶乘

```python
from math import factorial as factorial_math

factorial_math(3)
```
