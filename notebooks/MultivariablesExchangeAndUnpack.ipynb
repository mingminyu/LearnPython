# 解包(Unpack)

解包的过程中只能使用一个 `*var` 来接收不定长参数，熟悉 `*args` 和 `**kwargs` 的同学对这个一定不陌生。

```python
a, *b, c = (1, 2, 3, 4, 5)
print("a: ", a)
print("b: ", b)
print("c: ", c)
```

```python
a, *b, c = {1, 2, 3, 4, 5}
print("a: ", a)
print("b: ", b)
print("c: ", c)
```

如果是字典的话，解包的过程又是怎么样的呢？

```python
a, *b, c = {"a": 1, "b": 2, "c": 3, "d": 4}
print("a: ", a)
print("b: ", b)
print("c: ", c)
```

可以看到如果是字典的解包，实际上是用字段的键值作为解包的元素。
