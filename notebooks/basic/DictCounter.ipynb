# 对比性能

```python
from collections import Counter
from random import randint
from time import perf_counter

s = "abcdefghijklmnopqrst"

# 100_0000 表示 10,0000，下划线可以随便加，但是不能使用连续 2 个下划线
for i in range(10_0000):
    s += str(randint(0, 9))

t1 = perf_counter()

dic = {}

for char in s:
    if char not in dic:
        dic[char] = 1
    else:
        dic[char] += 1

t2 = perf_counter()
dic2 = Counter(s)
t3 = perf_counter()
print("User defined function Time: ", t2 - t1)
print("Counter Time: ", t3 - t2)
```
