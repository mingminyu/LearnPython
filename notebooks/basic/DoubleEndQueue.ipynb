```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

"""
    root(1)
   /       \
  n1(2)     n2(3)
 /     \
n3(4)   n4(5)
"""


root = TreeNode(1)
n1 = TreeNode(2)
n2 = TreeNode(3)
n3 = TreeNode(4)
n4= TreeNode(5)

root.left = n1
root.right = n2
n1.left = n3
n1.right = n4
```

```python
def dfs_preorder(node: TreeNode):
    """深度优先遍历(先序)"""
    if not node:
        return

    print(node.val, end=" ")
    dfs_preorder(node.left)
    dfs_preorder(node.right)


def dfs_medium_order(node: TreeNode):
    """深度优先遍历(中序)"""
    if not node:
        return

    dfs_medium_order(node.left)
    print(node.val, end=" ")
    dfs_medium_order(node.right)


def dfs_postorder(node: TreeNode):
    """深度优先遍历(后序)"""
    if not node:
        return

    dfs_postorder(node.left)
    dfs_postorder(node.right)
    print(node.val, end=" ")


print("先序遍历: ", end=" ")
dfs_preorder(root)

print("\n中序遍历: ", end=" ")
dfs_medium_order(root)

print("\n后序遍历: ", end=" ")
dfs_postorder(root)
```

```python
from collections import deque


def bfs():
    q = deque([(root, 0)])

    while q:
        node, level = q.popleft()

        if node is None:
            continue

        print(node.val, end=" ")
        q.append((node.left, level + 1))
        q.append((node.right, level + 1))


bfs()
```
