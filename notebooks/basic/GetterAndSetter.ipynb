```python
class Student:
    def __init__(self, name: str, score: int):
        self.name = name
        # 不想让实例访问该属性，可以使用双下划线
        self._score = score

    @property
    def score(self):
        return self._score

    @score.setter
    def score(self, n):
        if n > 90:
            self._score = 0.9 * n
        elif n > 80:
            self._score = 0.8 * n
        else:
            self._score = 0.6 * n


    @score.deleter
    def score(self):
        print("删除分数!")



stud = Student("Yu", 99)
stud.score
```
