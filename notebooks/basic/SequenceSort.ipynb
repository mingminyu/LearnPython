# 两两元素之间比较

上面的操作仅仅适用于依赖每个元素单个值进行排序，但是如果是比较方法是，先比较元素第一个值，然后再比较第二个值，这个自定义排序逻辑应该如何写？

```python
from functools import cmp_to_key

def cmp(a, b):
    print(a, b)

    if a[0] > b[0]:
        return 1
    elif a[0] < b[0]:
        return -1
    else:
        if a[1] > b[1]:
            return 1
        elif a[1] < b[1]:
            return -1
        else:
            return 0

lst5 = [(2, 2), (2, 1), (1, 3), (5, 2), (4, 1)]
lst5.sort(key=cmp_to_key(cmp))
lst5

# (2, 1) (2, 2) -1
# (1, 3) (2, 1) -1
# (5, 2) (1, 3) 1
# (5, 2) (2, 1) 1 => [      ,       ,    *1  ,   *2 ]
# (5, 2) (2, 2) 1 —> [(1, 3), (2, 1), (2, 2), (5, 2)]
# (4, 1) (2, 2) 1
# (4, 1) (5, 2) -1
```

```python
lst6 = [(2, 2), (2, 1), (1, 3), (5, 2), (1, 1)]
sorted(lst6, key=cmp_to_key(cmp))

# (2, 1) (2, 2) -1
# (1, 3) (2, 1) -1
# (5, 2) (1, 3) 1
# (5, 2) (2, 1) 1 => [      ,       ,    *1  ,   *2 ]
# (5, 2) (2, 2) 1 —> [(1, 3), (2, 1), (2, 2), (5, 2)]
# (1, 1) (2, 2) -1
# (1, 1) (2, 1) -1
# (1, 1) (1, 3) -1
```
