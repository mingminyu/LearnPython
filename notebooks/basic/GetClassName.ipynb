这个可以在类实现中使用，主要用于调用类成员属性

```python
class Student:
    def __init__(self, name):
        self.name = name

    def __repr__(self):
        return f"{self.__class__.__name__}('{self.name}')"


stud = Student(name="Alice")
stud
```
