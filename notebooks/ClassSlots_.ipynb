# 槽值 (slots) 的使用方法

```python
class C:
    ...


c = C()
c.f = 123
```

```python
import time
from typing import Union
```

```python
class A:
    __slots__ = ('f', 'g')

    def __call__(self, *args, **kwargs):
        ...


class B:
    def __call__(self, *args, **kwargs):
        ...


# noinspection PyPep8Naming
def timeit(className: Union[A, B]):
    p = className()
    start = time.time()

    for _ in range(10000000):
        p.f = 2

    print(className.__name__, time.time() - start)
```

```python
a = A()
a.f = 2
```

```python
a.c = 123
```

```python
timeit(A)
```

```python
timeit(B)
```
