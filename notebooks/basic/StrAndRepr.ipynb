```python
class User:
    def __init__(self, name: str, age: int):
        self.name = name
        self.age = age

    def __str__(self):
        return "Name: self.name / Age: self.age"

    def __repr__(self):
        return "Name: self.name \ Age: self.age"


user = User("li", 20)
user
```

```python
print(user)
```

在交互环境（Jupiter、IPython）中，以及手动调用 `repr` 函数，都会调用 `__repr__` 方法，使用 `print` 时则会调用 `__str__`
