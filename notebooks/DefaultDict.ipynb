使用 `collections.defaultdict`，我们可以实现更复杂数据结构，我们先来看一个示例：

```python
# 统计字符串中的字母个数
s = "aaabcdb"
dic1 = {}

for s_ in s:
    if s_ in dic1:
        dic1[s_].append(123)
    else:
        dic1[s_] = []

dic1
```

```python
from collections import defaultdict
```

使用 `defaultdict` 可以省略一步初始化过程

```python
dic2 = defaultdict(list)

dic2["a"] = 1
dic2["b"].append(123)
dic2
```

# 使用 `defaultdict` 实现一个图

```python
from random import randint
from collections import defaultdict
```

```python
graph = defaultdict(set)

for i in range(10):
    for _ in range(3):
        graph[i].add(randint(1, 10))

graph
```
